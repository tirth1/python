# File I/O

## opening a File

```python
f = open('test.txt', 'w')
```

```python
f = open('test.txt')  #equivalent to 'r'
```

## Closing a File

```python
f.close()
```

```python
try:
    f = open('test.txt')
finally:
    f.close()
```

## Writing to a File

```python
try:
    f = open('test.txt', 'w')
    f.write('This is a First line\n')
    f.write('This is a second line\n')
finally:
    f.close()
```

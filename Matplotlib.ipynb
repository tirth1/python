# Pyplot Tutorial:

```python
import matplotlib.pyplot as plt
```

```python
plt.plot([2, 4, 6, 4])
plt.ylabel("Numbers")
plt.xlabel("Indices")
plt.title("MyPlot")
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.xlabel("numbers")
plt.ylabel("sqr")
plt.grid()
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro')
plt.grid()

plt.show()
```

```python
import numpy as np

t = np.arange(0., 5., 0.2)

plt.plot(t, t**2, 'b--', label='^2')
plt.plot(t, t**2.2, 'rs', label='^2.2')
plt.plot(t, t**2.5, 'g^', label='^2.5')
plt.grid()
plt.legend()
plt.show()
```

## Controlling line properties

```python
x1 = [1, 2, 3, 4]
y1 = [1, 4, 9, 16]
x2 = [1, 2, 3, 4]
y2 = [2, 4, 6, 8]

lines = plt.plot(x1, y1, x2, y2)

#use keyword args
plt.setp(lines[0], color='r', linewidth=2.0)

#MATLAB style string value pairs
plt.setp(lines[1], 'color', 'g', 'linewidth', 2.0)

plt.grid()
```

```python
t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure(1)   #it will create 1 plot figure

plt.subplot(211)    #row, column, subplot_no.
plt.grid()
plt.plot(t1, np.cos(2*np.pi*t1), 'b-')


plt.subplot(212)    #row, column, subplot_no.
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')

plt.show()
```

```python
t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure(2)   #it will create 1 plot figure

plt.subplot(221)    #row, column, subplot_no.
plt.grid()
plt.plot(t1, np.cos(2*np.pi*t1), 'b-')


plt.subplot(222)    #row, column, subplot_no.
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')

plt.subplot(223)    #row, column, subplot_no.
plt.plot(t2, np.sin(2*np.pi*t2), 'b--')

plt.subplot(224)    #row, column, subplot_no.
plt.plot(t1, np.sin(2*np.pi*t1), 'r-')

plt.show()
```
